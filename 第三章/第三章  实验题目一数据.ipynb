```python
import requests
from bs4 import BeautifulSoup
res = requests.get('http://books.toscrape.com/')
html = res.text
soup = BeautifulSoup(html,'html.parser')
items = soup.find('ul',class_='nav').find('ul').find_all('li')
for item in items:
    kind = item.find('a')
    print(kind.text.strip(),'\n')
```
