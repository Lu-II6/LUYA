```python
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
dataset = {'ID':[1, 2, 3, 4, 5, 6],
        'Onion':[1, 0, 0, 1, 1, 1],
        'Potato':[1, 1, 0, 1, 1, 1],
        'Burger':[1, 1, 0, 0, 1, 1],
        'Milk':[0, 1, 1, 1, 0, 1],
        'Beer':[0, 0, 1, 0, 1, 0]}
```

```python
df = pd.DataFrame(dataset)
```

```python
df = df[['ID', 'Onion', 'Potato', 'Burger', 'Milk', 'Beer']]
```

```python
df
```

```python
frequent_itemsets = apriori(df[['Onion', 'Potato', 'Burger', 'Milk', 'Beer']], min_support=0.50, use_colnames=True)
```

```python
frequent_itemsets
```

```python
rules = association_rules(frequent_itemsets, num_itemsets=None, metric='lift', min_threshold=1)
rules = rules.sort_values(by="lift", ascending=False)
rules
```

```python
rules [(rules['lift'] > 1.125) & (rules['confidence'] > 0.8)]
```
