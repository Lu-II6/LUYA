```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import fpgrowth
from mlxtend.frequent_patterns import association_rules
import pyfpgrowth
import time
```

```python
with open('D:/桌面/basket.txt') as fp:
    products = [line.strip().split('\t') for line in fp]
onehot_enc = TransactionEncoder()
tmp_array = onehot_enc.fit_transform(products)
products_ = pd.DataFrame(tmp_array,columns=onehot_enc.columns_)
products_.head(5)
```

```python
t1 = time.time()
frequent_itemsets = fpgrowth(products_, min_support=0.1, use_colnames=True)
frequent_itemsets
```

```python
rules = association_rules(frequent_itemsets, num_itemsets=None, metric="lift", min_threshold=1)
rules[(rules['lift'] > 1) & (rules['confidence'] > 0.5)]
rules = rules.sort_values(by="lift", ascending=False)
rules
```

```python
print(time.time() - t1)
```
